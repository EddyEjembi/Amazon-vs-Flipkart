```python
import numpy as np
import pandas as pd
```

```python
amazon = pd.read_csv('amazon.csv')
amazon.head()
```

```python
flipkart = pd.read_csv('flipkart.csv')
flipkart.head()
```

```python
amazon.info()
```

```python
flipkart.info()
```

```python
amazon.rename(columns={'amazon_title': 'Title'}, inplace=True)
flipkart.rename(columns={'flipkart_title': 'Title'}, inplace=True)
amazon.head()
```

```python
flipkart.head()
```

```python
amazon.rename(columns={'amazon_isbn-10': 'isbn'}, inplace=True)
flipkart.rename(columns={'flipkart_isbn10': 'isbn'}, inplace=True)
amazon.head()
```

```python
flipkart.head()
```

```python
books = amazon.merge(flipkart, left_on='isbn', right_on='isbn', how='inner')
books.head(50)
```

```python
book = books.drop_duplicates()
book
```

```python
book.info()
```

### It was preferable to join with the isbn numbers

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.heatmap(book.isnull(), yticklabels=False, cbar=False)
```

```python
book.drop(['flipkart_author','Title_y'], axis=1, inplace=True)
book.rename(columns={'Title_x':'Title', 'amazon_author':'Author'}, inplace=True)
```

```python
book.dropna(inplace=True)
book.info()
```

```python
books = book[['Title', 'Author', 'isbn', 'amazon_rating', 'flipkart_stars', 'amazon_reviews count', 'flipkart_ratings count', 'amazon_price', 'flipkart_price']]
books
```

```python
books.to_excel('amazon vs flipkart.xlsx', index=False)
```
